```python
# импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import gamma
%matplotlib inline
```

### Центральная предельная теорема своими руками

В этом задании вам предстоит проверить работу центральной предельной теоремы, а также поработать с генерацией случайных чисел и построением графиков в Питоне.

Выберите ваше любимое непрерывное распределение (чем меньше оно будет похоже на нормальное, тем интереснее; попробуйте выбрать какое-нибудь распределение из тех, что мы не обсуждали в курсе). Сгенерируйте из него выборку объёма 1000, постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины (чтобы величины были в одном масштабе, не забудьте выставить у гистограммы значение параметра normed=true).

Ваша задача — оценить распределение выборочного среднего вашей случайной величины при разных объёмах выборок. Для этого при трёх и более значениях n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних. Используя информацию о среднем и дисперсии исходного распределения (её можно без труда найти в википедии), посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. Обратите внимание: для подсчёта значений этих параметров нужно использовать именно теоретические среднее и дисперсию вашей случайной величины, а не их выборочные оценки. Поверх каждой гистограммы нарисуйте плотность соответствующего нормального распределения (будьте внимательны с параметрами функции, она принимает на вход не дисперсию, а стандартное отклонение).

Опишите разницу между полученными распределениями при различных значениях n. Как меняется точность аппроксимации распределения выборочных средних нормальным с ростом n?

Решение должно представлять собой IPython-ноутбук, содержащий:
- код, генерирующий выборки и графики;
- краткие описания каждого блока кода, объясняющие, что он делает;
- необходимые графики (убедитесь, что на них подписаны оси);
- выкладки с вычислениями параметров нормальных распределений, аппроксимирующих выборочные средние при различных n;
- выводы по результатам выполнения задания.

```python
# задаем случайную величину, имеющую гамма-распределение и некоторые настраиваемые параметры a, d.
a = 4
d = 2
gamma_rv = sts.gamma(a, scale=d)

# генерируем выборку из 1000 случайных величин, распределенных по этому закону
sample = gamma_rv.rvs(size=1000)
```

```python
# строим график ее функции плотности
x = np.linspace(0, 30, 100)
pdf = gamma_rv.pdf(x)
plt.plot(x, pdf) 
plt.ylabel('$p(x)$')
plt.xlabel('$x$')
```

```python
# смотрим, насколько распределение нашей выборки sample, сгенерированной по гамма-распределению с параметрами (4, 2)
# отличается от теоретического идеального гамма-распределения с теми же параметрами
plt.hist(sample, bins=20, normed=True, histtype='stepfilled', alpha=0.5, color='lightgreen', label='sample')
plt.plot(x, gamma(a, scale=d).pdf(x), linewidth=2, label='expected pdf')
plt.legend(loc='upper right')
plt.ylabel('$p(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# видим, что не очень отличается (и в общем, оно и не должно сильно отличаться)
```

```python
# сделаем по 1000 выборок для шести разных n и положим их все в один большой массив samples
samp_sizes = [2, 5, 10, 25, 50, 100]
samples, samples_mean = [[], [], [], [], [], []],  [[], [], [], [], [], []]
new_mean = 0
# для каждого объема выборки: создаем выборку этого объема, добавляем ее в наш массив samples, в массив samples_mean добавляем
# сумму всех элементов этой выборки, делим на объем выборки. Повторяем это 1000 раз.
for i in range(len(samp_sizes)):
    size_curr = samp_sizes[i]
    for j in range(1000):
        samples[i].append(gamma_rv.rvs(size=size_curr))
        for k in range(size_curr):
            new_mean += samples[i][j][k]
        samples_mean[i].append(new_mean/size_curr)
        new_mean = 0
```

```python
# среднее и дисперсия для гамма-распределения с параметрами a и d равны ad и ad^2 соответственно (Википедия)
mean, var = gamma.stats(a, scale=d, moments='mv')
print(mean, var, sep='\n')
```

В дальнейшем выборочное среднее для выборки объема n будет обозначаться как $\overline{X_n}.\;$ 

```python
# Центральная предельная теорема утверждает, что выборочное среднее имеет распределение, близкое к N(mean, var/n)
# (при условии, что n стремится к бесконечности)
# теперь строим гистограммы для всех n по очереди, учитывая то, что scale - среднеквадратичное отклонение, а не дисперсия
n = samp_sizes[0]
norm_rv_0 = sts.norm(loc=mean, scale=np.sqrt(var/n))
plt.hist(samples_mean[0], bins=15, normed=True, histtype='stepfilled', alpha=0.5, color='lightgreen', label='sample_mean for n=2')
plt.plot(x, norm_rv_0.pdf(x), linewidth=2, label='pdf expected by Clt (n=2)')
plt.legend(loc='upper right')
plt.ylabel('$p(\overline{X}_n)$')
plt.xlabel('$\overline{X}_n$')
plt.show()
```

```python
# аналогично для n = 5:
n = samp_sizes[1]
norm_rv_1 = sts.norm(loc=mean, scale=np.sqrt(var/n))
plt.hist(samples_mean[1], bins=15, normed=True, histtype='stepfilled', alpha=0.5, color='lightgreen', label='sample_mean for n=5')
plt.plot(x, norm_rv_1.pdf(x), linewidth=2, label='pdf expected by Clt (n=5)')
plt.legend(loc='upper right')
plt.ylabel('$p(\overline{X}_n)$')
plt.xlabel('$\overline{X}_n$')
plt.show()
```

```python
# для n = 10:
n = samp_sizes[2]
norm_rv_2 = sts.norm(loc=mean, scale=np.sqrt(var/n))
plt.hist(samples_mean[2], bins=15, normed=True, histtype='stepfilled', alpha=0.5, color='lightgreen', label='sample_mean for n=10')
plt.plot(x, norm_rv_2.pdf(x), linewidth=2, label='pdf expected by Clt (n=10)')
plt.legend(loc='upper right')
plt.ylabel('$p(\overline{X}_n)$')
plt.xlabel('$\overline{X}_n$')
plt.show()
```

```python
# для n = 25:
n = samp_sizes[3]
norm_rv_3 = sts.norm(loc=mean, scale=np.sqrt(var/n))
plt.hist(samples_mean[3], bins=15, normed=True, histtype='stepfilled', alpha=0.5, color='lightgreen', label='sample_mean for n=25')
plt.plot(x, norm_rv_3.pdf(x), linewidth=2, label='pdf expected by Clt (n=25)')
plt.legend(loc='upper right')
plt.ylabel('$p(\overline{X}_n)$')
plt.xlabel('$\overline{X}_n$')
plt.show()
```

```python
# для n = 50:
n = samp_sizes[4]
norm_rv_4 = sts.norm(loc=mean, scale=np.sqrt(var/n))
plt.hist(samples_mean[4], bins=15, normed=True, histtype='stepfilled', alpha=0.5, color='lightgreen', label='sample_mean for n=50')
plt.plot(x, norm_rv_4.pdf(x), linewidth=2, label='pdf expected by Clt (n=50)')
plt.legend(loc='upper right')
plt.ylabel('$p(\overline{X}_n)$')
plt.xlabel('$\overline{X}_n$')
plt.show()
```

```python
# для n = 100:
n = samp_sizes[5]
norm_rv_5 = sts.norm(loc=mean, scale=np.sqrt(var/n))
plt.hist(samples_mean[5], bins=15, normed=True, histtype='stepfilled', alpha=0.5, color='lightgreen', label='sample_mean for n=100')
plt.plot(x, norm_rv_5.pdf(x), linewidth=2, label='pdf expected by Clt (n=100)')
plt.legend(loc='upper right')
plt.ylabel('$p(\overline{X}_n)$')
plt.xlabel('$\overline{X}_n$')
plt.show()
```

### Выводы:

Как видно из гистограмм, ЦПТ для наших выборок выполняется, и, как и ожидалось, при большем объеме выборки распределение для выборочного среднего лучше приближается нормальным распределением.

При n=25 уже достигается довольно высокая точность, а при n=2, n=5, и n=10 соответствующие по ЦПТ нормальные распределения недостаточно точно описывают выборочное среднее наших выборок.
